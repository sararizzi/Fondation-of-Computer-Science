```python
import pandas as pd # Pandas library to manipulate the dataframe 
from datetime import timedelta # datetime library to manage time
```

```python
trips = pd.read_csv('data.csv', dtype = {'store_and_fwd_flag': str}) # Importing the data with Pandas read_csv (), column 6 had mixed types so we converted the column type as object and filled null values as unknown    
trips.head()
```

```python
trips['store_and_fwd_flag'] = trips['store_and_fwd_flag'].fillna('Unknown') # column 6 had mixed types so we converted the column type as object and filled null values as unknown 
```

# 1) Extract all trips with trip_distance larger than 50

```python
dfifty = trips[trips['trip_distance'] > 50] # from the column 'trip_distance' we obtain all the rows with distance larger than 50
dfifty
```

# 2) Extract all trips where payment_type is missing

```python
trips[trips['payment_type'].isna()].head() # by using the pandas function .isna() we detect missing values in the column 'payment_type'
```

# 3) For each (PULocationID, DOLocationID) pair, determine the number of trips

```python
# we group by 'PULocationID', 'DOLocationID' using the function 'groupby' and the function .size() to obtain the number of trips for each group
# .reset_index(name = '') resets the index of the dataframe and names the new created column.
trips.groupby(['PULocationID', 'DOLocationID']).size().reset_index(name = 'Number of trips')
```

# 4) Save all rows with missing VendorID, passenger_count, store_and_fwd_flag, payment_type in a new dataframe called bad, and remove those rows from the original dataframe.

```python
# we use .isnull() and .any(axis = 1)(any row that satifies the condition) to create the dataframe with all trips having null values in the specified columns indicated
nullValue = trips[trips[['VendorID', 'passenger_count', 'store_and_fwd_flag', 'payment_type']].isnull().any(axis=1)]
nullValue
```

```python
# with the function .index we obtain the 'bad' rows index  and delete them from the original dataframe
trips = trips.drop(nullValue.index)
trips
```

#  5)  Add a duration column storing how long each trip has taken (use tpep_pickup_datetime, tpep_dropoff_datetime)

```python
# we substract the 'pickup_datetime' from the 'dropoff_datetime' to know how long each trip has taken
trips['duration'] = pd.to_datetime(trips['tpep_dropoff_datetime']) - pd.to_datetime(trips['tpep_pickup_datetime'])
```

```python
trips.head()
```

# 6) For each pickup location, determine how many trips have started there.

```python
# The same procedure as the 3rd point but grouping bu 'PULocationID' and naming the new created column 'trip_count'
trips.groupby('PULocationID').size().reset_index(name='trip_count') # if we want to count null values aswell
```

# 7) Cluster the pickup time of the day into 30-minute intervals (e.g. from 02:00 to 02:30)

```python
trips['parsed date'] = pd.to_datetime(trips['tpep_pickup_datetime'])
trips['minutes'] = trips['parsed date'].dt.hour * 60 + trips['parsed date'].dt.minute   #tranform time in minutes
trips['tpep_pickup_datetime_intervals'] = trips['minutes'] // 30      #dividing for 30 to create the slices
trips.head()
```

# 8) For each interval, determine the average number of passengers and the average fare amount.

```python
# we obtain the average number of passengers and the average fare amount by using .mean() and grouping by 'tpep_pickup_datetime_intervals'
interAvg = trips.groupby('tpep_pickup_datetime_intervals')[['passenger_count' , 'fare_amount']].mean().reset_index()
# we rename the columns using .rename() function
interAvg = interAvg.rename(columns={'passenger_count':'Avg_passenger','fare_amount':'Avg_fare_amount'})
interAvg.head()
```

# 9) For each payment type and each interval, determine the average fare amount

```python
# Same procedure as we did previously
tripsAvg = trips.groupby(['payment_type', 'tpep_pickup_datetime_intervals'])['fare_amount'].mean().reset_index(name='avg_fare_amount')
tripsAvg
```

# 10) For each payment type, determine the interval when the average fare amount is maximum

```python
# By using the previous dataframe, we group by 'payment_type' and get the id (interval) of the maximum 'avg_fare_amount
# In the dataframe result we store the corresponding rows through the ids 
maxid = tripsAvg.groupby('payment_type')['avg_fare_amount'].idxmax()
result = tripsAvg.loc[maxid]
result
```

# 11) For each payment type, determine the interval when the overall ratio between the tip and the fare amounts is maximum

```python
#add column ratio
trips['ratio'] = trips['tip_amount']/trips['fare_amount']

#remove row with nan or inf value
trips_cleaned = trips.dropna(subset=['ratio'])
trips_cleaned = trips_cleaned[~trips_cleaned['ratio'].isin([float('inf'), float('-inf')])]
trips_cleaned[['tip_amount', 'fare_amount', 'ratio']]
```

```python
#for each payment_type and intervals i found the max ratio
max_ratio = trips_cleaned.groupby(["payment_type" , "tpep_pickup_datetime_intervals"])["ratio"].mean().reset_index()
#find the interval with max ratio for each payment type
max_ratio = max_ratio.loc[max_ratio.groupby("payment_type")["ratio"].idxmax()]
max_ratio
```

# 13) Build a new dataframe (called common) where, for each pickup location we keep all trips to the 5 most common destinations (i.e. each pickup location can have different common destinations).

```python
#find for each cople (PULocationID, DOLocationID) the number of trip made
counter = trips.groupby(['PULocationID', 'DOLocationID']).size().reset_index(name='count').sort_values(by=['PULocationID','count'], ascending=[True, False])
counter
```

```python
# take for each PULocationID (groupby) the top 5 destination (head(5))
top_destinations = counter.groupby('PULocationID').head(5)
top_destinations.dropna()
```

```python
# Merge with original dataset to get the trips that has 'PULocationID' and 'DOLocationID' in the list of the top 5
common = trips.merge(top_destinations[['PULocationID', 'DOLocationID']], on=['PULocationID', 'DOLocationID'])
common 
```

# 14) On the common dataframe, for each payment type and each interval, determine the average fare amount

```python
# As done previously, we use the groupby function with the mean() function and name the newly created column as 'com_avg_fare'
commonAvg = common.groupby(['payment_type', 'tpep_pickup_datetime_intervals'])['fare_amount'].mean().reset_index(name='com_avg_fare')
commonAvg
```

# 15) Compute the difference of the average fare amount computed in the previous point with those computed at point 9.

```python
# To compute the difference we merge the two dataframes on 'payment_type' and 'tpep_pickup_datetime_intervals' 
mergeAvg = pd.merge(commonAvg, tripsAvg, on = ['payment_type', 'tpep_pickup_datetime_intervals'], how = 'inner')
# Calculate the difference of the two columns obtained
mergeAvg['diff_avg'] = mergeAvg['com_avg_fare'] - mergeAvg['avg_fare_amount']
mergeAvg
```

# 17) Build chains of trips. Two trips are consecutive in a chain if (a) they have the same VendorID, (b) the pickup location of the second trip  is also the dropoff location of the first trip, (c) the pickup time of the second trip is after the dropoff time of the first trip, and (d) the pickup time of the second trip is at most 2 minutes later than the dropoff time of the first trip.

```python
#keep only the needed columns and only the row without nan values
reduced = trips[['VendorID', 'PULocationID', 'DOLocationID', 'tpep_pickup_datetime', 'tpep_dropoff_datetime']]
reduced = reduced.dropna()

reduced['tpep_pickup_datetime'] = pd.to_datetime(reduced['tpep_pickup_datetime'])
reduced['tpep_dropoff_datetime'] = pd.to_datetime(reduced['tpep_dropoff_datetime'])

#I order the rows so that when I analyze a row on that all the previous ones have a lower starting time
reduced = reduced.sort_values(by=['VendorID', 'tpep_pickup_datetime'], ascending=[True, True])

reduced
```

```python
import time
from datetime import timedelta
import pandas as pd


#this function allows to find all the trips that can be precedents of current_trip
def find_valid_trips(reduced, current_trip, i):

    #search the valid trips only in the previous row (only in the rows that i have already analize)
    valid_trips = reduced.iloc[:i]
    #the past trips must certainly have arrived no more than two minutes earlier if we want a match (current_trip['tpep_pickup_datetime'] - timedelta(minutes=2) >= reduced['tpep_dropoff_datetime'])
    #if they arrived first there is no need to keep them in the database because going forward they will certainly not have matches (dataset is sorted by PU time)
    #with loc and with this condition i delete the unusefull rows (the one that has wrong DO time)
    reduced = reduced.loc[current_trip['tpep_pickup_datetime'] - timedelta(minutes=2) >= reduced['tpep_dropoff_datetime'],:]

    #search the trips that respect the conditions
    valid_trips = valid_trips[
    (current_trip['PULocationID'] == valid_trips['DOLocationID']) &
    (current_trip['tpep_pickup_datetime'] > valid_trips['tpep_dropoff_datetime']) &
    ((current_trip['tpep_pickup_datetime'] - valid_trips['tpep_dropoff_datetime']) < timedelta(minutes=2))
    ]

    #extract the indices of this trips
    valid_trip_indices = valid_trips.index.tolist()
    
    return valid_trip_indices


#this functions allows to update the dictionary with the trip sequences
def update_chain(chains_dict, valid_trips, current_trip):

    #for each trip identified as preceding the current trip, update the found sequences
    for past_trip in valid_trips:

        #if the index is already in the dictionary I update the chain that contains it 
        #it means that in the dictionary there is a sequence that ends with that journey
        if past_trip in chains_dict:
            #add current_trip at the existing chain
            chains_dict[past_trip].append(current_trip.name)
            #update the key, key contains the last trip of the chain
            chains_dict[current_trip.name] = chains_dict.pop(past_trip)
            return
        #if past trip isn't already in the chain i create a new chain
        else:
            #the new chain will contain the index of past trip followed by index of current_trip
            chains_dict[current_trip.name] = [past_trip, current_trip.name]



#start building the dictionary with the chains
#in the dictionary we have a sequence of trips and for each sequence the key is the last trip of the sequence (es. sequence [55, 67] has key = 67)
chains = {}
reduced = reduced.head(400000)

#for each vendorID
for vendor_id in reduced['VendorID'].unique():
    reduced = reduced[reduced['VendorID'] == vendor_id] 

    #analyze each row
    for i in range(len(reduced) - 1):

        #current trip is the the one in the row that i'm analyzing
        current_trip = reduced.iloc[i]

        #use the funcion to find all trips that may be the precedents of current_trip
        valid_trips = find_valid_trips(reduced, current_trip, i)

        #if we have some valid trip we need to uodate the found sequences
        if len(valid_trips)>0:
            #with this function we update the sequences in the dictionary
            update_chain(chains, valid_trips, current_trip)
                
#in chain i have a dictionary but I don't need it anymore the keys
all_chains = list(chains.values())
print('le sequenze trovate sono: ' , len(all_chains))
print('le sequenze sono: ', all_chains)

```

```python
from collections import defaultdict

number_to_sequences = defaultdict(list)

#find all the number in the sequences
for idx, sequence in enumerate(all_chains):
    for number in sequence:
        number_to_sequences[number].append(idx)

# fin the number that are in more than one sequence
duplicates = {num: seqs for num, seqs in number_to_sequences.items() if len(seqs) > 1}


if duplicates:
    for number, sequences in duplicates.items():
        print(f"Number {number} is in more than one sequence:  {[all_chains[i] for i in sequences]}")
else:
    print("no duplicates")
```
